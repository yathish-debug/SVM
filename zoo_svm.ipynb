```python
feature_df=df[['hair', 'feathers', 'eggs', 'milk', 'airborne', 'aquatic', #ignoring unnamed class_type column
       'predator', 'toothed', 'backbone', 'breathes', 'venomous', 'fins',
       'legs', 'tail', 'domestic', 'catsize']]
```

```python
feature_df # new dataframe
```

```python
x=(feature_df)#np.asarray
y=(df['class_type'])
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split #now splitting our data as per 80:20 rule
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=4)
```

```python
x_train.shape #to get the size 
y_train.shape
x_test.shape
y_test.shape
```

```python
from sklearn import svm #now building our model using support vector machines
classifier=svm.SVC(kernel='linear', gamma='auto', C=1) #using liner kernel trick,gamma as coeffecient for calcuation,c/penalty=1
classifier.fit(x_train,y_train) #training the model using x_train,y_train
y_predict=classifier.predict(x_test) #predicting x_test using the trained model
```

```python
def auto_accuracy(x,y): #this is a function for auto accuracy checker
    success=0
    for (a,b) in zip(x,y):
    #print(a,b)
        if (a==b):
            success+=1
            print(a,b ),print(success)
        else:
            print(a,b)
            print(success)
    
```

```python
auto_accuracy(y_test,y_predict) 
```
